```python
import os
from PIL import Image
import re
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
source_path = 'V3.dataset-skin-deasease-capstone/2validation'
folder = os.listdir(source_path)
folder
```

```python
clean_path = 'V3.dataset-skin-deasease-capstone/clean-dataset-validation'
for category in folder:
    os.mkdir(os.path.join(clean_path, category))
```

```python
folder2 = os.listdir(clean_path)
folder2
```

```python
for category in folder:
    source_path_per_category = os.path.join(source_path, category)
    clean_path_per_category = os.path.join(clean_path, category)
    list_files_per_category = os.listdir(source_path_per_category)
  #Resize and change format to .jpg
    for image in list_files_per_category:
        image_file = os.path.join(source_path_per_category, image)
        im = Image.open(image_file)
        #regex for select only number in filename
        number = re.search(r"^\S*", image)
        clean_file = number.group(0)
        #resize image and convert to .jpg
        file_save = os.path.join(clean_path_per_category, clean_file)
        im_final = im.resize((300, 300))
        im_final = im_final.convert("RGB")
        print(file_save)
        im_final.save(file_save)
```
